```python
def corre(data,threshold):
    corr_list = []
    ether = chaw.corr()
    for i in range(len(ether.columns)):
        for j in range(i):
            if abs(ether.iloc[i,j]) > threshold:
                col_name = ether.columns[i]
                corr_list.append(col_name)
    return corr_list
var = corre(chaw,0.6)
print(var)
```

## Splitting our data

```python
target_var = chaw['risk_flag']
variables = chaw.drop(['risk_flag'],axis = 1)
test_data = chin
```

## Resampling


```python
from imblearn.over_sampling import SMOTE
x_resample,y_resample = SMOTE().fit_sample(variables,target_var.values.ravel())
print(x_resample.shape)
print(y_resample.shape)
```

```python
print("Target_var Before Resampling:")
print(target_var.value_counts())
print("Target_var After Resmapling:")
y_resample = pd.DataFrame(y_resample)
print(y_resample[0].value_counts())
```

## Performing train_test_split 

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_resample,y_resample,test_size = 0.3, random_state = 0)
print("Shape of x_train:",x_train.shape)
print("shape of x_test:", x_test.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_test:", y_test.shape)

```

## Implementing Prediction Algorithm. 

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import cross_val_score
```

```python
X = normalised_x_train
Z = normalised_x_test
N = normalised_test_data
y_train_series = y_train.squeeze()
Y = y_train_series
y_test1 = y_test.squeeze()
```

```python
rf = RandomForestClassifier(n_estimators = 60)
rf_model = rf.fit(X,Y)
```

```python
print("The Training accuracy using Random Forest Classifier is:", rf_model.score(X,Y))
print("The Testing Accuracy using Random Forest Classifier is:", rf_model.score(Z,y_test1))
```

```python
predi = rf_model.predict(Z)
```

```python
predi
```

```python
print(confusion_matrix(y_test1,predi))
```

```python
print(classification_report(y_test1,predi))
```

```python
cm = confusion_matrix(y_test1,predi)
plt.rcParams['figure.figsize'] = (5,5)
plt.title("Confusion Matrix for Actual and Predicted", fontsize = 13)
sns.heatmap(cm, annot = True, cmap = 'Wistia', fmt = '.8g')
plt.xlabel('Actual', fontsize = 13)
plt.ylabel('Predicted', fontsize = 13)
plt.show()
```
